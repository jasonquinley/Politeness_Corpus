## We have parsing without errors here (sometimes happens). Now let's see how many columns there are. Let's build a list of the new file's columns from that. We'll give it a "C" for "classified". 

```python
COL_LIST=[]
for i in range(len(sf.columns)):
    COL_LIST.append('C'+str(i))

sf.head()
sf.columns=COL_LIST
sf.head()
```

### How can we separate these dictionaries into their own columns? 
#### Let's make a method to convert the keys and values into lists, then pull the first part of the list

```python
def KEY(dict_1):
    if dict_1 != None:
        try:
            return list(dict_1.keys())[0]
        except AttributeError:
            return 0
    else: 
        return 0
def VAL(dict_1):
    if dict_1 != None:
        try:
            return list(dict_1.values())[0]
        except AttributeError:
            return 0
    else: 
        return 0
```

```python


for j in range(len(sf.columns)):
    #print(sf['C'+str(j)])
    sample_file['S_'+str(j)]=pd.Series(sf['C'+str(j)].apply(KEY))
    sample_file['P_'+str(j)]= pd.Series(sf['C'+str(j)].apply(VAL))

```
